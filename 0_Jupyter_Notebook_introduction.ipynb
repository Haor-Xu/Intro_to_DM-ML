### Mode

If you made a mistake in your *Markdown* cell and have already executed it, you will notice that the cell can no longer be edited by simply selecting it. The reason is that you are in **Command Mode**. A Jupyter Notebook has two different modes:

1. **Edit Mode**: Allows you to change the contents of the cell. 

2. **Command Mode**: Allows to modify the cell as a whole and use keyboard shortcuts. The content of the cell cannot be changed.

You can switch between the two modes by either pressing <kbd>ESC</kbd> and <kbd>Enter</kbd>, or by clicking outside or inside the cell (for Markdown cells you have to double-click). You can always tell which mode you are in. If the selected cell has a green border you are in **Edit Mode**, if the cell has a blue border you are in **Command Mode**. Try it out.

### Keyboard shortcuts and other tricks

#### Command Mode Shortcuts

There are a number of useful keyboard shortcuts in `Command Mode` to use Jupyter Notebook more efficiently and easily. Remember to switch between `Command Mode` and `Edit Mode` with <kbd>Esc</kbd> and <kbd>Enter</kbd>.

<kbd>m</kbd>: Convert cell to Markdown.

<kbd>y</kbd>: Convert cell to code.

<kbd>D</kbd>+<kbd>D</kbd>: Delete cell.

<kbd>Shift</kbd>+<kbd>Arrow up/Arrow down</kbd>: Select multiple cells. Once the cells are selected, you can work with them as with a batch, e.g. execute, copy, delete, etc.

<kbd>Shift</kbd>+<kbd>M</kbd>: Merge selected cells.

<kbd>Shift</kbd>+<kbd>Tab</kbd>: [press once] Displays the parameters to pass to a function. <br />
<kbd>Shift</kbd>+<kbd>Tab</kbd>: [press 3 times] Displays additional information about the method.

Under "Help - Keyboard Shortcuts" you can find more shortcuts.

#### Line Magics

Line magics are functions that can be executed in cell mode. As argument they get the rest of the line in which they are executed. Line Magics are executed by placing a '%' character in front of the command. Some of the most useful line magics are:

`%matplotlib inline`: This command ensures that all plots from matplotlib are displayed in the output cell. Also, this ensures that the graphs are saved when the notebook is saved.

`%reload_ext autoreload`, `%autoreload 2`: Reloads all modules before executing a new line. If a module has been edited, it is no longer necessary to execute the corresponding import commands again, as they are updated automatically.

The three commands just presented should always be executed at the beginning of a notebook.

```python
%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

`%timeit`: Executes a line 10,000 times and displays the average runtime required.

```python
%timeit [i+1 for i in range(1000)]
```

`%debug`: Allows to examine a function with incorrect execution, based on the [Python Debugger](https://docs.python.org/3/library/pdb.html).

```python
for i in range(1000):
    a = i+1
    b = 'string'
    c = b+1
```

```python
%debug
```

The commands for operating the debugger can be found [here](https://docs.python.org/3/library/pdb.html).
