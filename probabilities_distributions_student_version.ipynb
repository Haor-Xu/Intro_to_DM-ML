Let's assume that we have a dataset consists of the values from the measurement of sizes of the grains in a specimen.

```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(seed=42)
grain_size = rng.normal(loc=25, scale=5, size=1000) # With size being equal to 1000, we measure the size of 1000 different grains
```

1. Print the first 10 outcomes of the sample space.

2. Comment on why one cannot directly compute the probability that a randomly selected grain has a size of 15.5 $\mu m$?

3. Plot a histogram of the outcomes and comment on how we can utilize it about the probabilities.


### Answers of Exercise 2

```python
for i in range(10):
    print(grain_size[i])
```

Because we don't have the range of the random variable, we can't say that the grain size is normally distributed. 
  
  
We can only say that the grain size of the 1000 grains we measured is normally distributed.

```python
plt.hist(grain_size, bins=100, edgecolor='black')
hide = plt.show()

#answer
fig, ax = plt.subplots(figsize=(5, 3.5), dpi=125)

count, bins, ignored = plt.hist(grain_size, bins=100, edgecolor='black', density=True)
ax.set(xlabel='Grain size distribution',ylabel='number of grains')
```
